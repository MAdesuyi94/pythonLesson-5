## Functions

## What is a function?
- ### A function is a group of code that performs a specific task
- ### Functions help to organize code, make it reusable, and improve readability

## Example of a function use case:
- ### A function that greets you with your name. "Greetings Matthew!!!"

## What is modularization?
- ### Modularization is the process of dividing a program into separate functions

## Why should I modularize my code?
- ### It makes the code easier to read and maintain
- ### It allows for code reuse
- ### It helps to isolate and fix bugs
- ### It allows us to divide a program into different part
- ### It makes it easier for us to plan and develop different parts of a program
- ### It makes it easier for collaboration

## How do I define and call a function?
- ### To define a function, we use the `def` keyword followed by the function name and parentheses `()`. The code block within every function starts with a colon `:` and is indented.

```python
def greet(name):
    print("Greetings " + name + "!!!")
```

- ### To call a function, we simply use the function name followed by parentheses `()` and pass any required arguments.

```python
greet("Matthew")
```

- ### The function name should describe the task that the function is trying to do. In the above example, the function named `greet` should actually greet a person.
- ### The code block under the function definition is a series of statements that will run everytime the function is called.
- ### The variable within the function name is called an argument. 
    - ### An argument is a value that is passed to the function when it is called. In the above example, the argument is "Matthew".
    - ### The argument is used within the function to perform the task. In the above example, the argument is used to greet the person with their name.

````python
def say_hello(name):
    print(f"Hello {name}!!!!")

say_hello("Matthew")
````

````python
def say_hello(first_name, last_name):
    print(f"Hello {first_name} {last_name}!!!!")

say_hello("Matthew", "Adesuyi")
````

## Top-down design
- ### Top-down design is a programming approach that involves breaking down a complex problem into smaller, more manageable parts.
- ### It involves starting with the main function and then breaking it down into smaller functions that perform specific tasks.
- ### This approach helps to organize code, improve readability, and make it easier to debug and maintain.

## Hierarchy Chart
- ### A hierarchy chart is a visual representation of the top-down design of a program.
- ### It shows the main function at the top and the smaller functions that it calls below it.
- ### Each function is represented by a box, and the arrows indicate the flow of control between the functions.

![Function_Hierarchy_Chart](images/Function_Hierarchy_Chart.png)

````python
def greet(name):
    print(f"Greetings {name}!!!!")

def display_temperature(temperature):
    print(f"The temperature is {temperature} degrees.")

def show_sky_color(sky_color):
    print(f"The sky is {sky_color}.")

def main():
    greet("Alice")
    display_temperature(72)
    show_sky_color("blue")

main()
````

## Local Variables
- ### A local variable is a variable that is defined within a function and can only be accessed within that function.

```python
def greet(name):
    message = "Greetings " + name + "!!!"  # message is a local variable
    print(message)

greet("Matthew")
```

- ### In the above example, the variable `message` is a local variable because it is defined within the `greet` function. It cannot be accessed outside of the function.
- ### Local variables are useful because they help to prevent naming conflicts and keep the code organized.


````python
def greet(name):
    # Local Variable
    greeting_message = f"Greetings {name}!!!!"
    print(greeting_message)

greet("Alice")
````

## Argument
- ### An argument is a value that is passed to a function when it is called.
- ### Argument as also known as a parameter variable.

```python
def greet(name):  # name is an argument
    print("Greetings " + name + "!!!")

greet("Matthew")  # "Matthew" is the argument value
```

## Keyword Argument
- ### A keyword argument is an argument that is passed to a function using the name of the parameter.

```python
def greet(name, message):
    print(message + " " + name + "!!!")

greet(name="Matthew", message="Greetings")
```
- ### In the above example, the arguments are passed using the names of the parameters. This allows us to pass the arguments in any order.
- ### You can either put your function arguments by position (order of the arguments), or you can use named arguments.
- ### You can also use a mixture of both, but positional (unnamed) arguments must come first.

````python
def greet(name):
    print(f"Greetings {name}!!!!")

greet(name="Alice")
````

## Global Variable
- ### A global variable is a variable that is defined outside of any function and can be accessed from anywhere in the program.

```python
message = "Hello, World!"  # message is a global variable

def greet():
    print(message)

greet()
print(message)
```

````python
name = "Alice" # Global Variable

def greet():
    print(f"Greetings {name}!!!!")

greet()
````
